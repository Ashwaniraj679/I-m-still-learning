```python
df = pd.read_csv('Holdings.csv')
```

```python
updated_columns = ['Security', 'No. of Smallcases', 'Quantity', 'Average Cost',
       'Portfolio Weight %', 'LTP', 'Invested Value', 'Current Value',
       'P & L', 'Net Change %', 'Daily Change', 'Daily Change %']
df.columns = updated_columns
```

```python
df= df.drop('No. of Smallcases',axis = 1)
df.head()
```

```python
temp = df[["Security" ,"Quantity", "Invested Value", "Current Value"]].copy()
```

```python
temp.sort_values(["Quantity"], ascending = False).reset_index(drop =True)
```

```python
total_amount_invested = temp[["Invested Value"]].sum()
current_invested_value = temp[["Current Value"]].sum()
total_amount_invested , current_invested_value  
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the bar plot
plt.bar(temp['Security'], temp['Current Value'], label='Current Value')
plt.bar(temp['Security'], temp['Invested Value'], label='Invested Value')

# Set the title and axis labels
plt.title('Investment Values by Security')
plt.xlabel('Security')
plt.ylabel('Value')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
