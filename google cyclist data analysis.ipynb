```python
df['started_at'] = pd.to_datetime(df['started_at'], format='%Y-%m-%d %H:%M:%S')
df['ended_at'] = pd.to_datetime(df['ended_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
print(df['started_at'].dtype)
print(df['started_at'])
```

```python
df = df.assign(ride_length=df['ended_at'] - df['started_at'])

# Display the DataFrame
print(df)
```

```python
timedelta = pd.Timedelta('7 days 00:00:00')
df_filtered = df.loc[df['ride_length'] > timedelta]
df_filtered
```

```python
# Get a summary of the 'column1' column
summary = df['ride_length'].describe()

# Print the mean
print(summary)
```

```python
index_to_drop = df_filtered.index
df.drop(index_to_drop, inplace=True)
```

Converting Member for 1 and casual for 0


```python
df['member_casual'].replace(['member', 'casual'], [0, 1], inplace=True)
```

```python
# Create a bar chart of the casual and member values
df['member_casual'].value_counts().plot(kind='bar')

# Add labels and title
plt.xlabel('Member or Casual')
plt.ylabel('Count')
plt.title('Count of Member and Casual')

# Show the chart
plt.show()
```

```python
df['member_casual'].value_counts()
```

Create a column called “day_of_week,” and calculate the day of the week that each ride strts. Formating as General or as a number with no decimals, noting that
1 = Sunday and 7 = Saturday

```python
# Assume the DataFrame has a column called 'start_time'
df['day_of_week'] = df['started_at'].dt.weekday
df['day_of_week'] = df['day_of_week'].apply(lambda x: x+1)
df['day_of_week'] = df['day_of_week'].apply(lambda x: 7 if x == 0 else x)
```

```python
df['ride_length'].describe()
```

Creating the line plot to figure out ride length on particular day of week

```python
df.groupby('day_of_week')['ride_length'].mean().plot(kind='line')
plt.xlabel('Day of Week')
plt.ylabel('Ride Length')
plt.title('Ride Length Line plot by Day of Week')
plt.show()
```

```python
df_grouped = df.groupby(['member_casual','day_of_week'])['ride_length'].mean().reset_index()
df_grouped
```

```python
df_groupe = df_grouped.pivot(index='day_of_week', columns='member_casual', values='ride_length')
df_groupe
```
