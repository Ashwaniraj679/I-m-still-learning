```python
df = pd.DataFrame(holds['data']) 
```

```python
df['total_price'] = df['availableQty'] * df['avgCostPrice']
df
```

```python
df  = df.sort_values(by= 'total_price', ascending = False)
# create a bar chart
ax = df.plot(kind='barh',x='tradingSymbol', y='total_price')

# set chart title and axis labels
ax.set_title('Total Price by Trading Symbol')
ax.set_xlabel('Trading Symbol')
ax.set_ylabel('Total Price')
plt.show()
```
